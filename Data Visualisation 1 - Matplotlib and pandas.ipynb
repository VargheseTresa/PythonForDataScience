```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import randn, randint, uniform, sample
```

```python
#create a dataframe and time series with random values
df = pd.DataFrame(randn(1000), index = pd.date_range('2019-06-07', periods = 1000), columns=['value'])
ts = pd.Series(randn(1000), index = pd.date_range('2019-06-07', periods = 1000))
print(df.head())

#perform cumulative sum to make the charts more meaningful
ts = ts.cumsum()
df['value'] = df['value'].cumsum()
df.head()
```

### Load data using seaborn

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

#### Setting labels and title

```python
ax = iris.plot(figsize=(15,8), title='Iris Dataset')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
```

## Bar plot

```python
#x-axis = indexes
#y-axis = values in columns
```

```python
ts.plot(kind = 'bar')
```

```python
#remove categorical data
df = iris.drop(['species'], axis = 1)
df.head()
```

```python
#bar plot on single row of df
df.iloc[0].plot(kind='bar')
```

```python
#another way of creating bar plot
df.iloc[0].plot.bar()
```

```python
#load titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Histogram

```python
#x-axis = range of the data
#y-axis = frequncies of occurance of data
```

```python
iris.head()
```

```python
iris.plot.hist()
#non numeric columns are automatically excluded
```

```python
#Another way
iris.plot(kind = 'hist')
```

```python
iris.plot(kind = 'hist', stacked = False, bins = 100)
```

### horizontal histogram

```python
iris.plot(kind = 'hist', stacked = True, bins = 50, orientation = 'horizontal')
```

```python
iris['sepal_width'].plot(kind = 'hist', stacked = True, bins = 50)
```

```python
iris['sepal_width'].diff()
```

```python
iris['sepal_width'].diff().plot(kind = 'hist', stacked = True, bins = 50)
```

### Formatting the histogram

```python
#alpha specifies the transparency
df = iris.drop(['species'], axis = 1)
df.diff().hist(color = 'g', alpha = 0.2, figsize=(10,10))
```

## Scatter Plot
