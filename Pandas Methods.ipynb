# Merging two data frames

```python
# Creating data frames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[0, 1, 2, 3])
df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8,9,10,11])
df3
```

## Appending dataframes along rows
### concat()

```python
#axis = 0 for rows
df_cat1 = pd.concat([df1,df2,df3], axis=0)
print("\nAfter concatenation along row\n",'-'*30, sep='')
print(df_cat1)

#selecting values from merged data frame
df_cat1.loc[2]
```

```python
df_cat1.iloc[[2]]
```

## Merging data frame along columns

```python
df_cat2 = pd.concat([df1,df2,df3], axis=1)
#rows are combined based on index. 
#If no value found in a particular index, NaN is placed
df_cat2
```

```python
#Replace NaN with value 0
df_cat2.fillna(value=0, inplace=True)

print("\nAfter filling missing values with zero\n",'-'*60, sep='')
print(df_cat2)
```

## Join operation on dataframes
### inner join

```python
#create data frames

left = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
print("Data frame Left","-"*30, left, sep="\n", )
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
print("\nData frame Right","-"*30, right, sep="\n", )
```

```python
merge1= pd.merge(left,right,how='inner',on='key')

print("\nAfter simple merging with 'inner' method\n",'-'*50, sep='')
merge1
```

```python
# Example without matching keys on both data frame

left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
print("Data frame Left","-"*30, left, sep="\n", )
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K5', 'K6'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
print("\nData frame Right","-"*30, right, sep="\n", )
```

```python
pd.merge(left, right, how='inner', on='key1')
#here only rows with matching keys are joined, others are eliminated
```

```python
#Example with multiple keys
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
print("Data frame Left","-"*30, left, sep="\n", )
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
print("\nData frame Right","-"*30, right, sep="\n", )
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

### left and right joins

```python
pd.merge(left, right, how='left',on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right',on=['key1', 'key2'])
```

## join operators

```python

left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 
print("Data frame Left","-"*30, left, sep="\n", )

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
print("\nData frame Right","-"*30, right, sep="\n", )
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
